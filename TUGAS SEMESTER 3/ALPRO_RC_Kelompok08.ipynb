A. Unvectorized Solution

```python
import numpy as np 
N = 10000
dart_locations = 2 * np.random.rand(N, 2) - 1
```

```python
Ncircle = [0]
for s, d in dart_locations:
    if np.sqrt (s**2 + d**2) < 1:
        Ncircle.append(Ncircle[-1] + 1)
    else:
        Ncircle.append (Ncircle[-1])
```

```python
running_presupposition = []
for n_total, n_circle in enumerate(Ncircle [1:]):
    running_presupposition.append(4*n_circle / (n_total +1))
```

B. Vertorized Solution

```python
N = 10000
dart_locations = 2 *np.random.rand(N, 2) - 1
```

```python
dist_out_of_origin = np.sqrt((dart_locations**2).sum(axis = 1))
```

```python
is_in_circle = dist_out_of_origin < 1
```

```python
num_in_circle = np.cumsum(is_in_circle)
num_thrown = np.arange(1, N+1)
```

```python
running_presupposition = 4 * num_in_circle / num_thrown
```

A. Unvectorized Solution

```python
import numpy as np
predicted = np.array([[1,0,1,0],
         [2,1,0,1],
         [-1,3,5,1],
         [1,1,1,0],
         [0,4,3,1]])
actual = np.array([[1,0,1,0],
        [2,1,0,1],
        [-1,3,5,0],
        [1,1,1,0],
        [0,4,3,1]])
predicted_class = [predicted[0,3:],predicted[1,3:],predicted[2,3:],predicted[3,3:],predicted[4,3:]]
actual_class = [actual[0,3:],actual[1,3:],actual[2,3:],actual[3,3:],actual[4,3:]]
i = 0
TP = 0
TN = 0
while i<5:
        if(predicted_class[i] == 1 and actual_class[i] == 1):
            TP = TP+1
        elif(predicted_class[i] == 0 and actual_class[i] == 0):
            TN = TN+1
        i = i+1
N = len(predicted)
accu = (TP+TN)/N #unvectorized
print ("Akurasi dari model Neural Network adalah dengan cara unvectorized solution : ", accu)
```

B. Vectorized Solution

```python
import numpy as np
predicted = np.array([[1,0,1,0],
         [2,1,0,1],
         [-1,3,5,1],
         [1,1,1,0],
         [0,4,3,1]])
actual = np.array([[1,0,1,0],
        [2,1,0,1],
        [-1,3,5,0],
        [1,1,1,0],
        [0,4,3,1]])
predicted_class = [predicted[0,3:],predicted[1,3:],predicted[2,3:],predicted[3,3:],predicted[4,3:]]
actual_class = [actual[0,3:],actual[1,3:],actual[2,3:],actual[3,3:],actual[4,3:]]
i = 0
TP = 0
TN = 0
while i<5:
        if(predicted_class[i] == 1 and actual_class[i] == 1):
            TP = TP+1
        elif(predicted_class[i] == 0 and actual_class[i] == 0):
            TN = TN+1
        i = i+1
N = len(predicted)
accuracy = np.divide(np.add(TP,TN), N) #vectorized
print ("Akurasi dari model Neural Network adalah dengan cara vectorized solution : ",accuracy)
```
